# IP Address 관리 체계

- node와 node간에 데이타 패킷을 전송하기 위해서는 각 node에 고유 주소를 필요로 하는데 이때 그 고유 주소가 IP입니다.

- IP주소는 32bit를 가지는데 8bit씩 끊어서 4자리로 나누게 되고 각 자리는 " . " 로 구분된다.
  ex) 210.211.194.2
- 32bit 의 크기를 가지므로 2^32 만큼의 인터페이스에 고유한 주소를 부여할수 있게 된다
- 1 byte(:12) 씩 4자리로 나누는 이유는 4byte 숫자보다 인간이 인지하기가 쉽기 때문이다.
- 현재 모든 인터페이스에 고유한 주소를 부여하기에는 턱없이 부족한 숫자인데, 이를 해결하기 위해서 128bit 주소체계를 가지는 IPv6(:12)가 이미 나와있다.

# TCP (Transmission Control Protocol)
- IP를 사용하는 프로토콜이며 데이터를 여러개의 패킷으로 나누고 도착지에서는 완전한 데이터로 패킷을 재조립 한다.
- 서버와 클라이언트간에 데이타를 신뢰성있게 전달하기 위해 만들어진 프로토콜이다.

### 특징
- IP위에서 동작하도록 설계 되어있다.
- TCP는 상위층이 넘겨준 데이터를 세그먼트라는 단위로 쪼개어 가공하고 하위층으로 넘겨준다
- 세그먼트에 순서를 부여하여 전송,수신하여 순서가 뒤바뀌는 일이 없도록 하고 뒤바뀌는 경우 순서를 재조합 한다
- TCP는 손실을 검색해내서, 이를 교정한다.
  (네트워크선로를 통해 전달되는 과정에서 손실되거나 순서가 뒤바뀌어서 전달될수 있기때문에)
- 데이타를 전송하기 전에 데이타전송을 위한 연결을 만드는 연결지향 프로토콜이다.
  (다른 말로 신뢰성 스트림 서비스라고 부릅니다)
